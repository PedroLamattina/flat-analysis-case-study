12. ¿Hay alguna tendencia en el valor medio de las propiedades en los distintos meses del año 2018? ¿Y en el valor máximo o mínimo?

```python
sSQL = """
SELECT 
    period AS mes,
    ROUND(AVG(price), 2) AS precio_medio,
    MAX(price) AS precio_maximo,
    MIN(price) AS precio_minimo
FROM 
    madrid
GROUP BY 
    period
ORDER BY 
    mes;
"""

pd.read_sql(sSQL, conn)
```

16. Calcula el precio promedio de las propiedades nuevas y de las de segunda mano en buen estado, y 
muestra ambos precios.

```python
sSQL = """
SELECT 
'Pisos Obra Nueva' AS tipo,
ROUND(AVG(price), 2) AS precio_promedio
FROM madrid
WHERE BUILTTYPEID_1 = 1

UNION ALL

SELECT 
'Segunda Mano (Buen Estado)' AS tipo,
ROUND(AVG(price), 2) AS precio_promedio
FROM madrid
WHERE BUILTTYPEID_3 = 1


"""

pd.read_sql(sSQL, conn)
```

17. Calcula el número total de propiedades que tienen terraza y piscina, y cuántas de ellas también 
tienen ascensor. <br/> Utiliza la función CASE para contar las propiedades con estas combinaciones. 

```python
sSQL = """
SELECT
    SUM(CASE WHEN HASTERRACE = 1 AND HASSWIMMINGPOOL = 1 THEN 1 ELSE 0 END) AS terraza_piscina,
    SUM(CASE WHEN HASTERRACE = 1 AND HASSWIMMINGPOOL = 1 AND HASLIFT THEN 1 ELSE 0 END) AS terraza_piscina_ascensor
FROM madrid


"""

pd.read_sql(sSQL, conn)
```

```python
sSQL = """
SELECT
    COUNT(CASE WHEN HASTERRACE = 1 AND HASSWIMMINGPOOL = 1 THEN 1 ELSE NULL END) AS terraza_piscina,
    COUNT(CASE WHEN HASTERRACE = 1 AND HASSWIMMINGPOOL = 1 AND HASLIFT = 1 THEN 1 ELSE NULL END) AS terraza_piscina_ascensor
FROM madrid
"""

pd.read_sql(sSQL, conn)
```
